```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
dataframe = pd.read_csv('train.csv')
dataframe
```

```python
df = dataframe.copy()
```

1. Drop Id column since it's the same with row index

```python
df = df.drop(columns = ['Id'], axis=1)
df
```

2. Change Home Owership column to number

```python
df['Home Ownership'].unique()
```

```python
replace_dic = {'Own Home': 0, 'Home Mortgage': 1, 'Rent': 2, 'Have Mortgage': 3}
df['Home Ownership'].replace(replace_dic, inplace=True)
df
```

3. Fill missing value by mean for Annual Income column.

```python
df['Annual Income'].isna().sum()
```

```python
df['Annual Income'] = df['Annual Income'].fillna(df['Annual Income'].mean())
df
```

4. Convert Years in current job column to number and fill missing value

```python
df['Years in current job'] = df['Years in current job'].str.extract(pat='(\d+)', expand=False)
df['Years in current job'] = df['Years in current job'].astype(float)
df['Years in current job'] = df['Years in current job'].fillna(df['Years in current job'].mean())
df
```

5. Drop Tax Liens column since 95% of them are zero

```python
df = df.drop(columns = ['Tax Liens'], axis=1)
```

6. Fill missing value for column Months since last delinquent

```python
df['Months since last delinquent'] = df['Months since last delinquent'].fillna(0.0)
df
```

7. Fill missing value for column Bankruptcies

```python
df['Bankruptcies'] = df['Bankruptcies'].fillna(0.0)
```

8. Convert column Purpose to number

```python
key_arr = df['Purpose'].unique()
replace_dic_2 = {}
for i in range(len(key_arr)):
    replace_dic_2[key_arr[i]] = float(i)
print(replace_dic_2)
df['Purpose'].replace(replace_dic_2, inplace=True)    
df
```

```python
print(df['Term'].unique())
replace_dic_3 = {'Short Term': 0.0, 'Long Term': 1.0}
df['Term'].replace(replace_dic_3, inplace=True)
df
```

```python
print(df['Credit Score'].isna().sum())
df['Credit Score'] = df['Credit Score'].fillna(df['Credit Score'].mean())
df
```
